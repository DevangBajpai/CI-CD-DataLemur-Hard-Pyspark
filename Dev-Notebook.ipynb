```python
schema = StructType([
    StructField("id", IntegerType()),
    StructField("dt", StringType())
])

data = [
[725 ,'03/03/2022 12:00:00'],
[245 ,'03/28/2022 12:00:00'],
[112 ,'03/05/2022 12:00:00'],
[245 ,'04/29/2022 12:00:00'],
[123 ,'05/01/2022 12:00:00'],
[245 ,'05/05/2022 12:00:00'],
[725 ,'05/25/2022 12:00:00'],
[311 ,'06/06/2022 12:00:00'],
[112 ,'06/07/2022 12:00:00'],
[123 ,'02/22/2022 12:00:00'],
[112 ,'03/15/2022 12:00:00'],
[245 ,'07/25/2022 12:00:00']
]

w_s = Window.partitionBy(col("id")).orderBy(col("mth"))

df = spark.createDataFrame(data, schema)
df.withColumn("dt", F.to_timestamp("dt", "MM/dd/yyyy HH:mm:ss"))\
  .withColumn("mth", month(col("dt")))\
  .withColumn("id_2", F.lag(col("id")).over(w_s))\
  .withColumn("mth2", F.lag(col("mth")).over(w_s))\
  .filter(col("mth2").isNull() | ((col("mth") - col("mth2")) != 1))\
  .groupBy(col("mth"))\
  .agg(
      F.countDistinct(col("id")).alias("cnt") 
  )\
  .display()
```

_Senior Managers_

```python
schema = StructType([
    StructField("emp_id", IntegerType()),
    StructField("manager_id", IntegerType()),
    StructField("name", StringType())]
)

data = [
(1	,101	,'Duyen'),
(101	,1001	,'Rick'),
(103	,1001	,'Rick'),
(2	,102	,'Delton'),
(102	,1002	,'Jason'),
(8	,107	,'Tina'),
(107	,1004	,'Tom'),
(1004	,1002	,'Jason'),
(1004	,1014	,'Brandon'),
(125	,1004	,'Tom'),
(126	,1004	,'Tom'),
(127	,1004	,'Tom'),
(128	,1004	,'Tom'),
(129	,1004	,'Tom'),
(12	,150	,'Raman'),
(150	,1001	,'Rick')
]

df = spark.createDataFrame(data, schema)

df_e = df.alias("e")
df_m = df.alias("m")

manager_list = (
    df_e.join(df_m, df_e.emp_id == df_m.manager_id, "inner")
    .select(
        df_e.emp_id, df_e.manager_id, df_e.name
    ).distinct()
)

df_manger_left = manager_list.alias("manger_left")
df_manger_right = manager_list.alias("manger_right")

result = (
    df_manger_left.join(
        df_manger_right,
        how="left",
        on = df_manger_left.emp_id == df_manger_right.manager_id
    )
    .groupBy(
        df_manger_left.manager_id,
        df_manger_left.name
    )
    .agg(
        F.count(df_manger_left.emp_id).alias("direct_reportees"),
        F.count(df_manger_right.manager_id).alias("cnt_manager")
    )
)

result.filter(col("cnt_manager")  == 0).select("name", "direct_reportees").display()


#  1 job -- display
#  4 stages -- join, distinct, join, groupby
```

_Advertiser Status_

```python
adv_s = StructType([
    StructField('id', StringType()),
    StructField('status', StringType()),
])

adv_d = [
('bing', 'NEW'),
('yahoo', 'NEW'),
('alibaba', 'EXISTING'),
('baidu', 'EXISTING'),
('target', 'CHURN'),
('tesla', 'CHURN'),
('morgan', 'RESURRECT'),
('chase', 'RESURRECT')
]

paid_s = StructType([
    StructField('id', StringType()),
    StructField('paid', IntegerType()),
])

paid_d = [
('yahoo', 45.00),
('alibaba', 100.00),
('target', 13.00),
('morgan', 600.00),
('fitdata', 25.00)
]

df_adv = spark.createDataFrame(adv_d, schema=adv_s)
df_paid = spark.createDataFrame(paid_d, schema=paid_s)

df_adv.join(df_paid, how = 'full_outer', on="id")\
      .select(
              F.coalesce(df_adv.id, df_paid.id).alias('id'), 
              F.coalesce(df_adv.status, F.lit('NEW_USER')).alias('status'), 
              df_paid.paid)\
      .withColumn('new_status',
              when(col("paid").isNull() & col("status").isin('NEW','EXISTING','CHURN','RESURRECT'), 'CHURN')
              .when(col("paid").isNotNull() & col("status").isin('NEW','EXISTING','RESURRECT'), 'EXISTING')
              .when(col("paid").isNotNull() & col("status").isin('CHURN'), 'RESURRECT').otherwise(F.lit('NEW')))\
      .select('id', 'new_status')\
      .orderBy('id')\
      .display()

#1 job #3 stages
```
